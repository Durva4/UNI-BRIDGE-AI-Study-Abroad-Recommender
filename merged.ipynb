```python
import pandas as pd

# loading both datasets
qs = pd.read_csv("qs_clean.csv")
cost = pd.read_csv("cost_country.csv")
```

```python
qs
```

```python
qs_country = qs.groupby("Country").agg({
    "Academic_Reputation": "mean",
    "Employer_Reputation": "mean",
    "Research_Quality": "mean",
    "Employment_Outcomes": "mean"
}).reset_index()
```

```python
df = pd.merge(qs_country, cost, on="Country", how="inner")
```
