```python
#selecting only required columns from the perspective of a student
cost_df = df[[
    "Country",
    "Level",
    "Tuition_USD",
    "Living_Cost_Index",
    "Rent_USD"
]]
```

```python
#aggregating cost data at the country level using mean values to represent the average cost per country.
cost_country = cost_df.groupby("Country").agg({
    "Tuition_USD": "mean",
    "Living_Cost_Index": "mean",
    "Rent_USD": "mean"
}).reset_index()
```
